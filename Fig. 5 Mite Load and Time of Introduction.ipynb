```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from symengine import exp, cos, pi, Abs
from jitcdde import y, t, jitcdde

# Common parameters
phi_B     = 0.018
phi_H     = 0.007
epsilon_B = 2000
b         = 500
k         = 5000
a         = 500
tau       = 12
gamma_B   = 0.003
gamma_H   = 0.05
delta_1   = 0.06
delta_2   = 0.06
epsilon_T = 1
gamma_T   = 1 / 24
theta     = 105
kappa     = 1 / 9

# Initial colony conditions
B0_base = 5000
H0_base = 10000
f0_base = 1000

# Simulation setup
DT  = 0.02
eps = 1e-6
lam = 120

# Collapse threshold
H_crit = 1000

# Grid
T_vals = np.arange(0, 101, 10)
V_vals = np.arange(0, 101, 10)

line_color = (230 / 255, 97 / 255, 0 / 255)

base = plt.colormaps["viridis"]
viridis_light = LinearSegmentedColormap.from_list(
    "viridis_light", base(np.linspace(0.22, 1.00, 256))
)

def build_and_run(T0, V0, chi, sigma, starttime, stoptime, B0=B0_base, H0=H0_base, f0=f0_base):
    q0 = delta_1 * T0 / (a + B0) + delta_2 * V0 / (a + B0)
    I0 = q0 * tau

    def pos(x):
        return (x + Abs(x)) / 2

    IDX_B, IDX_H, IDX_T, IDX_V, IDX_f, IDX_I = 0, 1, 2, 3, 4, 5

    B  = y(IDX_B)
    H  = y(IDX_H)
    Tm = y(IDX_T)
    V  = y(IDX_V)
    f  = y(IDX_f)
    I  = y(IDX_I)

    B_tau = y(IDX_B, t - tau)
    T_tau = y(IDX_T, t - tau)
    V_tau = y(IDX_V, t - tau)

    B_pos, H_pos, T_pos = pos(B), pos(H), pos(Tm)
    V_pos, f_pos, I_pos = pos(V), pos(f), pos(I)
    B_tau_pos, T_tau_pos, V_tau_pos = pos(B_tau), pos(T_tau), pos(V_tau)

    S = (f_pos / (b + f_pos)) * (H_pos / (k + H_pos))

    infestation = delta_1 * (B_pos / (a + B_pos)) * T_pos
    infection   = delta_2 * (B_pos / (a + B_pos)) * V_pos

    q_now = delta_1 * T_pos / (a + B_pos) + delta_2 * V_pos / (a + B_pos)
    q_tau = delta_1 * T_tau_pos / (a + B_tau_pos) + delta_2 * V_tau_pos / (a + B_tau_pos)
    dI    = q_now - q_tau

    exp_B = exp(-gamma_B * tau - I_pos)
    Phi = chi + sigma * cos(2 * pi * (t - theta) / 365)

    dB = epsilon_B * S - infestation - infection - gamma_B * B_pos - kappa * exp_B * B_tau_pos
    dH = kappa * exp_B * B_tau_pos - gamma_H * H_pos
    dT = epsilon_T * infestation - gamma_T * T_pos
    dV = epsilon_T * infection   - gamma_T * V_pos
    df = Phi * (H_pos / (k + H_pos)) - phi_B * B_pos - phi_H * H_pos

    dB += lam * (B_pos - B)
    dH += lam * (H_pos - H)
    dT += lam * (T_pos - Tm)
    dV += lam * (V_pos - V)
    df += lam * (f_pos - f)
    dI += lam * (I_pos - I)

    DDE = jitcdde([dB, dH, dT, dV, df, dI])

    def history(_s):
        return [B0, H0, T0, V0, f0, I0]

    DDE.past_from_function(history)
    DDE.step_on_discontinuities()
    DDE.set_integration_parameters(
        atol=1e-8, rtol=1e-6, min_step=1e-12, max_step=DT, first_step=DT
    )

    t0 = float(DDE.t)
    start = max(starttime + eps, t0)
    times = np.arange(start, stoptime + DT / 2, DT)

    minH = H0
    t_collapse = np.nan

    for tt in times:
        state = DDE.integrate(float(tt))
        Hval = float(state[IDX_H])

        if Hval < minH:
            minH = Hval

        if np.isnan(t_collapse) and (Hval < H_crit):
            t_collapse = float(tt)
            break

    return minH, t_collapse

def compute_maps(chi, sigma, day0_shift):
    starttime = day0_shift
    stoptime = 365 + day0_shift if day0_shift > 0 else 365

    minH_map = np.zeros((len(V_vals), len(T_vals)))
    tc_map   = np.full((len(V_vals), len(T_vals)), np.nan)

    for iV, V0 in enumerate(V_vals):
        for iT, T0 in enumerate(T_vals):
            minH, tc = build_and_run(
                T0=float(T0),
                V0=float(V0),
                chi=chi,
                sigma=sigma,
                starttime=starttime,
                stoptime=stoptime,
            )
            minH_map[iV, iT] = minH
            tc_map[iV, iT] = tc

    return minH_map, tc_map

def draw_panel(ax, minH_map, panel_label, xticks, yticks):
    nV, nT = minH_map.shape
    T_edges = np.linspace(T_vals[0], T_vals[-1], nT + 1)
    V_edges = np.linspace(V_vals[0], V_vals[-1], nV + 1)

    T_centers = 0.5 * (T_edges[:-1] + T_edges[1:])
    V_centers = 0.5 * (V_edges[:-1] + V_edges[1:])
    TTc, VVc = np.meshgrid(T_centers, V_centers)

    im = ax.pcolormesh(
        T_edges,
        V_edges,
        minH_map,
        shading="flat",
        cmap=viridis_light,
        edgecolors=(1, 1, 1, 0.75),
        linewidth=0.6,
        antialiased=False,
        vmin=1000,
        vmax=10000,
        zorder=1,
    )

    ax.contour(
        TTc,
        VVc,
        minH_map,
        levels=[H_crit],
        colors=[line_color],
        linewidths=2.5,
        zorder=10,
    )

    ax.set_xlim(T_vals[0], T_vals[-1])
    ax.set_ylim(V_vals[0], V_vals[-1])
    ax.margins(x=0, y=0)

    ax.set_xticks(xticks)
    ax.set_yticks(yticks)
    ax.tick_params(axis="both", labelsize=16, width=1.6, length=6, colors="black")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.text(-0.13, 1.12, panel_label, transform=ax.transAxes, color="black", fontsize=18, va="top", ha="left")
    return im

panels = [
    {"label": "(a)", "chi": 600, "sigma": 300, "day0_shift": 0,   "xticks": [0, 50, 100], "yticks": [0, 50, 100]},
    {"label": "(b)", "chi": 600, "sigma": 300, "day0_shift": 105, "xticks": [0, 30, 100], "yticks": [0, 30, 100]},
    {"label": "(c)", "chi": 600, "sigma": 300, "day0_shift": 287, "xticks": [0, 10, 100], "yticks": [0, 10, 100]},
    {"label": "(d)", "chi": 500, "sigma": 400, "day0_shift": 0,   "xticks": [0, 50, 100], "yticks": [0, 50, 100]},
    {"label": "(e)", "chi": 500, "sigma": 400, "day0_shift": 105, "xticks": [0, 40, 100], "yticks": [0, 40, 100]},
    {"label": "(f)", "chi": 500, "sigma": 400, "day0_shift": 287, "xticks": [0, 10, 100], "yticks": [0, 10, 100]},
]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

last_im = None

for ax, panel in zip(axes, panels):
    minH_map, tc_map = compute_maps(panel["chi"], panel["sigma"], panel["day0_shift"])

    print(f'Approximate threshold for {panel["label"]}:')
    collapse_map = np.isfinite(tc_map)
    for iT, T0 in enumerate(T_vals):
        idx = np.where(collapse_map[:, iT])[0]
        if len(idx) == 0:
            print(f"T0={T0:>4}: no collapse up to V0={V_vals[-1]}")
        else:
            V_thr = V_vals[idx[0]]
            print(f"T0={T0:>4}: V0 ~ {V_thr}")

    last_im = draw_panel(
        ax,
        minH_map,
        panel["label"],
        panel["xticks"],
        panel["yticks"],
    )

for ax in axes[:]:
    ax.set_xlabel("Initial number of virus-free mites, $T(0)$", fontsize=15)
    ax.set_ylabel("Initial number of \n virus-carrying mites, $V(0)$", fontsize=15)

fig.subplots_adjust(right=0.88, wspace=0.45, hspace=0.35)

cbar = fig.colorbar(last_im, ax=axes, location="right", fraction=0.035, pad=0.02)
cbar.set_label("Minimum hive bee population", fontsize=18)
cbar.ax.tick_params(labelsize=16, width=1.6, length=6, colors="black")
cbar.set_ticks([1000, 5000, 10000])
cbar.set_ticklabels(["1,000", "5,000", "10,000"])

plt.savefig("PRCC.png", dpi=300, bbox_inches="tight")
plt.show()
```
