```python
## FIGURE 1 (PHI)

import numpy as np
import matplotlib.pyplot as plt

# Time array (1 year, daily)
t = np.arange(0, 366)  # days 0 to 365

# Function
chi       = 600
sigma     = 300
theta     = 105  
Phi =  chi + sigma * np.cos(2 * np.pi * (t-theta) / 365)

# Plot
Phi_color = (51/255, 117/255, 56/255)
plt.figure(figsize=(10, 5))
plt.plot(t, Phi, label="$\\Phi$", linestyle='--', color =Phi_color, linewidth=4,)
plt.xlabel("Time (days)", color='black', fontsize=15, labelpad=10, ha='center', fontweight='bold')
plt.ylabel("Forage supply (in kg)", color='black', fontsize=15, labelpad=10, ha='center', fontweight='bold')
plt.tick_params(axis='x', which='major', labelsize=18, width=2, length=6, colors='black')
plt.tick_params(axis='y', which='major', labelsize=18, width=1.6, length=6, colors='black')

ax = plt.gca()
#ax.set_yticks([960, 600, 320])
ax.set_yticks([900, 600, 300])
ax.set_yticklabels(['0.9', '0.6', '0.3'])
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.33), fontsize=18, frameon=False, ncol=1)
plt.grid(False)
plt.xlim(0, 365)

# Month ends (2025, non-leap year)
month_ends = [31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]
month_labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Add vertical lines for month boundaries
for me in month_ends:
    plt.axvline(me, color='gray', linestyle='--', alpha=0.3)

# Add vertical line in start of January
jan_start = [0]
for me in jan_start:
    plt.axvline(me, color='gray', linestyle='--', alpha=0.3)

# Place month labels at the middle of each month
month_starts = [1] + [x+1 for x in month_ends[:-1]]
month_mids = [(s+e)//2 for s,e in zip(month_starts, month_ends)]
plt.xticks(month_mids, month_labels)

plt.savefig("Phi.png", dpi=300, bbox_inches='tight')  # saves as PNG with high resolution
plt.show()
```
